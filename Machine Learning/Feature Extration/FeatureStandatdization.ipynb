# Feature Engineering 


1) standardization - z-score

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
list(df['total_bill'])
```

```python
sns.histplot(df['total_bill'])
```

```python
total_bill=list(df['total_bill'])
```

```python
import numpy as np
```

```python
mean=np.mean(total_bill)
std=np.std(total_bill)
```

```python
mean , std
```

```python
normalized_data=[]
for i in total_bill:
    z_score=(i-mean)/std
    normalized_data.append(z_score)

print(normalized_data)    
```

```python
display(normalized_data)    
```

```python
sns.histplot(normalized_data)
```

1) sklearn StandardScaler 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler
```

```python
scaler.fit(df[['total_bill','tip']])
```

```python
scaler.transform(df[['total_bill','tip']])
```

```python
pd.DataFrame(scaler.transform(df[['total_bill','tip']]), columns=['total_bill','tip'])
```

```python
df1=pd.DataFrame(scaler.transform(df[['total_bill','tip']]), columns=['total_bill','tip'])
```
