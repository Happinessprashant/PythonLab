### SMOTE (SYNTHETIC MINORITY OVERSAMPLING TECHNIQUE)

####SMOTE is a technique used in machine learning to address imnlanced dataset where the minority class has significantly fewer instance than the class . SMOTE involves generating synthetic instance of the minority class by interpolating between existing instance

#09/10/2023

```python
from sklearn.datasets import make_classification
```

```python
make_classification(n_samples=1000,n_features=2,n_redundant=0,n_clusters_per_class=1 , weights=[0.90],random_state=1)
```

```python
## x is independent
## y is dependent

x,y=make_classification(n_samples=1000,n_features=2,n_redundant=0,n_clusters_per_class=1 , weights=[0.90],random_state=1)
```

```python
x,y
```

```python
import pandas as pd
df1=pd.DataFrame(x,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1,df2],axis=1)
```

```python
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'],final_df['f2'] , c=final_df  ['target'])
```

```python
## transform the dataset
oversampling=SMOTE()
oversampling.fit_resample(final_df[['f1','f2']] , final_df  ['target'])
X,y=oversampling.fit_resample(final_df[['f1','f2']] , final_df  ['target'])
```
